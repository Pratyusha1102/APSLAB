## Fractional Knapsack

```python
def fractional_knapsack(values, weights, W):
  n = len(values)

  ratios = []
  for i in range(n):
    ratio = values[i] / weights[i]
    ratios.append((ratio, i))

  ratios.sort(reverse=True)

  total_value = 0
  remaining_capacity = W

  print(f"Knapsack Capacity = {W}")
  print("Items sorted by value to weight ratio:")
  for ratio, i in ratios:
    print(f"Items {i+1}: Value = {values[i]}, Weight = {weights[i]}, Ratio = {ratio: .2f}")

  print("\n")

  for ratio, i in ratios:
    if remaining_capacity == 0:
      break
    if weights[i] <= remaining_capacity:
      total_value += values[i]
      remaining_capacity -= weights[i]
      print(f"Take full item {i+1} (Value = {values[i]}, Weight = {weights[i]})")
    else:
      fraction = remaining_capacity / weights[i]
      total_value += values[i] * fraction
      print(f"Take {fraction: .2f} fraction of Item {i+1} (Value added={values[i]*fraction:.2f}, Weight used={remaining_capacity})")
      remaining_capacity = 0

  print(f"Total value in knapsack = {total_value:.2f}")
  return total_value
```

```python
values = [60, 100, 120]
weights = [10, 20, 40]
W = 50

fractional_knapsack(values, weights, W)
```

## Huffman Coding

```python
import heapq
```

```python
class Node:
  def __init__(self, char, freq):
    self.char = char
    self.freq = freq
    self.left = None
    self.right = None

  def __lt__(self, other):
    return self.freq < other.freq

def huffman_encoding(chars, freqs):
  heap = [Node(chars[i], freqs[i]) for i in range(len(chars))]
  heapq.heapify(heap)

  while len(heap) > 1:
    left = heapq.heappop(heap)
    right = heapq.heappop(heap)
    new_node = Node(None, left.freq + right.freq)
    new_node.left = left
    new_node.right = right
    heapq.heappush(heap, new_node)

  root = heap[0]

  codes = {}
  def generate_codes(node, code=""):
      if node is None:
          return
      if node.char is not None:
          codes[node.char] = code
      generate_codes(node.left, code + "0")
      generate_codes(node.right, code + "1")

  generate_codes(root)
  return codes
```

```python
chars = ['A', 'B', 'C', 'D', 'E', 'F']
freqs = [5, 9, 12, 13, 16, 45]

codes = huffman_encoding(chars, freqs)

print("Huffman Codes:")
for char in codes:
    print(f"{char}: {codes[char]}")
```
