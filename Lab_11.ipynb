# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')

        for next_city in range(n):
            if mask & (1<<next_city) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1<<next_city))
                if new_cost < ans:
                    ans = new_cost

        dp[city][mask] = ans
        return ans

    return visit(0, 1)
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

print("Minimum travelling cost:", tsp(graph))
```

### 1. Print path of cities visited

To print the path, we need to modify the original `tsp` function slightly to keep track of the chosen next city at each step. Since you asked not to modify the existing `tsp` function, I'll create a new `tsp_with_path` function based on the same logic.

```python
def tsp_with_path(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]
    path_trace = [[-1]*(1<<n) for _ in range(n)] # To store the next city in the optimal path

    def visit(city, mask):
        if mask == (1<<n) - 1: # All cities visited, return to start
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        min_cost = float('inf')
        best_next_city = -1

        for next_city in range(n):
            # Check if next_city has not been visited yet
            if not (mask & (1<<next_city)):
                current_cost = graph[city][next_city] + visit(next_city, mask | (1<<next_city))
                if current_cost < min_cost:
                    min_cost = current_cost
                    best_next_city = next_city

        dp[city][mask] = min_cost
        path_trace[city][mask] = best_next_city # Store the best next city for this state
        return min_cost

    # Calculate the minimum cost starting from city 0, with only city 0 visited
    min_total_cost = visit(0, 1)

    # Reconstruct the path
    path = [0] # Start at city 0
    current_city = 0
    current_mask = 1 # City 0 is visited

    while current_mask != (1<<n) - 1:
        next_hop = path_trace[current_city][current_mask]
        path.append(next_hop)
        current_city = next_hop
        current_mask |= (1<<current_city) # Mark next_hop as visited

    path.append(0) # Return to the starting city

    return min_total_cost, path

# Test with the existing graph
min_cost_existing, path_existing = tsp_with_path(graph)
print(f"Minimum travelling cost for existing graph: {min_cost_existing}")
print(f"Path taken for existing graph: {path_existing}")
```

### 2. Take graph input from user and 3. Test with different number of cities

This section will allow you to define your own graph by entering the number of cities and their distances. It will then use the `tsp_with_path` function to find and print the optimal path and cost for your custom graph.

```python
def get_user_graph():
    while True:
        try:
            n = int(input("Enter the number of cities (n, must be > 1): "))
            if n <= 1:
                print("Number of cities must be greater than 1. Please try again.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter an integer.")

    user_graph = []
    print(f"\nEnter the distance matrix for your {n} cities. Each row should have {n} space-separated integers.")
    print("For example, for 3 cities, you might enter: `0 10 15` then `10 0 20` then `15 20 0`")
    print("Remember: the distance from a city to itself must be 0 (e.g., graph[i][i] = 0).")

    for i in range(n):
        while True:
            try:
                row_input = input(f"Enter {n} distances for city {i} (space-separated): ").split()
                if len(row_input) != n:
                    print(f"Please enter exactly {n} values. Try again.")
                    continue

                row_values = [int(val) for val in row_input]

                # Basic validation: diagonal must be 0
                if row_values[i] != 0:
                    print(f"Distance from city {i} to itself (diagonal) must be 0. Please re-enter row.")
                    continue

                user_graph.append(row_values)
                break # Exit inner loop if row is valid
            except ValueError:
                print("Invalid input. Please enter integers for distances. Try again.")

    return user_graph

user_graph = get_user_graph()
print("\nYour entered graph:")
for row in user_graph:
    print(row)

# Calculate TSP for the user-defined graph
min_cost_user, path_user = tsp_with_path(user_graph)
print(f"\nMinimum travelling cost for user-defined graph: {min_cost_user}")
print(f"Path taken for user-defined graph: {path_user}")
```
