## **Counting Sort** ##

```python
def counting_sort(arr):
    if not arr:
        return arr

    max_val = max(arr)
    min_val = min(arr)

    range_of_elements = max_val - min_val + 1
    count = [0] * range_of_elements
    output = [0] * len(arr)

    # Count occurrences
    for num in arr:
        count[num - min_val] += 1

    # Cumulative count
    for i in range(1, len(count)):
        count[i] += count[i - 1]

    # Build output array (stable)
    for num in reversed(arr):
        output[count[num - min_val] - 1] = num
        count[num - min_val] -= 1

    return output
```

```python
input_list = [4,3,2,1,4,3,2,4,3,4]
sorted_list = counting_sort(input_list)
print("Sorted Output: ",sorted_list)
```

## **Bucket Sort** ##

```python
def bucket_sort(arr):
    if len(arr) == 0:
        return arr

    min_val = min(arr)
    max_val = max(arr)
    n = len(arr)

    # Number of buckets
    bucket_count = n

    # Range of each bucket
    bucket_range = (max_val - min_val + 1) / bucket_count

    # Create empty buckets
    buckets = [[] for _ in range(bucket_count)]

    # Distribute elements into buckets
    for num in arr:
        index = int((num - min_val) / bucket_range)
        if index == bucket_count:  # edge case for max value
            index -= 1
        buckets[index].append(num)

    # Sort each bucket
    for bucket in buckets:
        bucket.sort()

    # Concatenate buckets
    sorted_array = []
    for bucket in buckets:
        sorted_array.extend(bucket)

    return sorted_array
```

```python
arr = [8,5,3,1,9,6,0,7,4,2,5]
print("Sorted Output:", bucket_sort(arr))
```

## **Radix Sort** ##

```python
def counting_sort(arr, exp):
    n = len(arr)
    output = [0] * n
    count = [0] * 10  # digits 0-9

    # Count occurrences of digits
    for i in range(n):
        index = (arr[i] // exp) % 10
        count[index] += 1

    # Cumulative count
    for i in range(1, 10):
        count[i] += count[i - 1]

    # Build output array (stable)
    for i in range(n - 1, -1, -1):
        index = (arr[i] // exp) % 10
        output[count[index] - 1] = arr[i]
        count[index] -= 1

    # Copy back
    for i in range(n):
        arr[i] = output[i]


def radix_sort(arr):
    max_val = max(arr)
    exp = 1

    while max_val // exp > 0:
        counting_sort(arr, exp)
        exp *= 10

    return arr
```

```python
arr = [18,5,100,3,1,19,6,0,7,4,2]
print("Sorted Output:", radix_sort(arr))
```
