# MATRIX CHAIN MULTIPLICATION Dynamic Programming with table filling.
Problem Description:
Find minimum number of scalar multiplications needed to multiply a chain of matrices..

#Theory
Cost of multiplying two matrices:
If A is p×q and B is q×r, cost = p × q × r

Dynamic Programming idea:
Break problem into smaller subproblems.
Store results in a table to avoid recomputation.

m[i][j] = minimum cost to multiply Ai to Aj

## Algorithm
1. Take array p[] of size n
2. Create table m[n][n]
3. Set m[i][i] = 0
4. Fill table diagonally
5. For each chain length L:
    Compute m[i][j]
    Try all possible k between i and j
6. Choose minimum cost
7. Return m[1][n-1]

```python
import sys

def matrix_chain_multiplication(p):
    n = len(p)

    # Create DP table for costs
    m = [[0 for _ in range(n)] for _ in range(n)]
    # Create DP table for split points (optimal parenthesization)
    s = [[0 for _ in range(n)] for _ in range(n)]

    # L is chain length
    for L in range(2, n):
        for i in range(1, n - L + 1):
            j = i + L - 1
            m[i][j] = sys.maxsize

            for k in range(i, j):
                cost = (m[i][k] +
                        m[k+1][j] +
                        p[i-1] * p[k] * p[j])

                if cost < m[i][j]:
                    m[i][j] = cost
                    s[i][j] = k  # Store the split point

    return m, s
```

```python
def print_optimal_parenthesization(s, i, j):
    if i == j:
        return f"A{i}"
    else:
        left = print_optimal_parenthesization(s, i, s[i][j])
        right = print_optimal_parenthesization(s, s[i][j] + 1, j)
        return f"({left}{right})"

def naive_multiplication_cost(p):
    n = len(p)
    if n <= 2: # Single matrix or two matrices
        return 0 if n == 1 else p[0] * p[1] * p[2] # A1*A2

    cost = 0
    current_dims = [p[0], p[1]] # Dimensions of the first matrix (A1)

    # Multiply A1 * A2, then (A1*A2) * A3, and so on.
    for k in range(2, n):
        # p[k-1] and p[k] are dimensions for A_k
        # current_dims[0] x current_dims[1] is the result of previous multiplications
        # we are multiplying (current_result) by A_k, which has dimensions p[k-1] x p[k]
        cost += current_dims[0] * current_dims[1] * p[k]
        current_dims[1] = p[k] # The new resulting matrix will have dimensions current_dims[0] x p[k]
    return cost
```

```python
# Given p for the exercise
p_exercise = [5, 10, 3, 12, 5, 50, 6]

# Compute minimum cost and split points
m_table, s_table = matrix_chain_multiplication(p_exercise)

# Minimum cost
n_exercise = len(p_exercise)
min_cost = m_table[1][n_exercise - 1]
print(f"Minimum multiplication cost: {min_cost}")

# Optimal parenthesization
optimal_paren = print_optimal_parenthesization(s_table, 1, n_exercise - 1)
print(f"Optimal parenthesization: {optimal_paren}")

# Naive multiplication cost (left-to-right)
naive_cost = naive_multiplication_cost(p_exercise)
print(f"Naive (left-to-right) multiplication cost: {naive_cost}")

# Compare number of multiplications saved
saved_multiplications = naive_cost - min_cost
print(f"Number of multiplications saved: {saved_multiplications}")
```

```python
# Input dimensions
p = [10, 30, 5, 60]

table = matrix_chain_multiplication(p)

print("Minimum multiplication cost:", table[1][len(p)-1])

print("\nDP Table:")
for row in table[1:]:
    print(row[1:])
```
